```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sb
%matplotlib inline
bdata = pd.read_csv('/content/brake test readings.csv')
print(bdata.shape)
bdata.head(10)
```

```python
bdata.drop(["Unnamed: 6"], axis = 1, inplace = True)
bdata
```

```python
bdata['Torque'] = ''
bdata['Pout'] = ''
bdata['Pin'] = ''
bdata['%efficiency'] = ''
bdata.to_csv("/content/brake test readings.csv", index=False)
print(bdata.shape)
bdata.head()
```

```python
for i in range(9): 
  bdata.loc[i,'Torque']= 9.81*(bdata.loc[i,'w1']-bdata.loc[i,'w2'])*bdata.loc[i,'R(cm)']
  bdata.loc[i,'Pin']=bdata.loc[i,'V(volts)']*bdata.loc[i,'I(amps)']
  bdata.loc[i,'Pout']=(6.2831*bdata.loc[i,'N']*bdata.loc[i,'Torque'])/60
  bdata.loc[i,'%efficiency']=bdata.loc[i,'Pout']/bdata.loc[i,'Pin']*100
print(bdata) 
```

```python
x = px.line(bdata, x = 'Pout', y = 'Torque', title='Output vs Torque')
x.show()
y= px.line(bdata, x = 'Pout', y = 'N', title='Output vs Speed')
y.show()
z = px.line(bdata, x = 'Pout', y = 'I(amps)', title='Output vs load current')
z.show()
a = px.line(bdata, x = 'Pout', y = '%efficiency', title='Output vs %efficiency')
a.show()
```
